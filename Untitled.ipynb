```python
#435(3)

import numpy as np
from matplotlib import pyplot as plt
x=np.linspace(-3,3)


x1=np.repeat(2,50)
y1=np.linspace(-20,10)

x2=np.repeat(-3,50)
y2=np.linspace(-20,10)



#print(x)
plt.grid()
plt.xlim(-4,4)
plt.ylim(-20,10)

#y=x*x*x-9*x
y=-x*x*x-3*x*x+5


plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)

plt.show()



```
